```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load dataset
df = pd.read_csv("best-selling game consoles.csv")

#scatter plot for Units Sold vs Release Year
plt.figure(figsize=(10, 6))
plt.scatter(df['Released Year'], df['Units sold (million)'], alpha=0.5)
plt.title('Units Sold vs. Released Year')
plt.xlabel('Released Year')
plt.ylabel('Units Sold (million)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
total_units_sold = df.groupby('Console Name')['Units sold (million)'].sum().reset_index()

#sorts dataFrame by total units, decreaseing order
total_units_sold = total_units_sold.sort_values(by='Units sold (million)', ascending=False)

#bar graph for Total Units Sold by Console
plt.figure(figsize=(12, 8))
plt.bar(total_units_sold['Console Name'], total_units_sold['Units sold (million)'], color='skyblue')
plt.title('Total Units Sold by Console')
plt.xlabel('Console Name')
plt.ylabel('Total Units Sold (million)')
plt.xticks(rotation=45, ha='right') #rotate x-axis labels to read easier
plt.tight_layout()
plt.show()
```

```python
Q3 = "Which console type sold the most? Which console type sold the least?"
print(Q3)
```

```python
#group data by console type, calculate total units sold
console_type_sales = df.groupby('Type')['Units sold (million)'].sum().reset_index()

#bar graph, total units sold by console type
plt.figure(figsize=(10, 6))
plt.bar(console_type_sales['Type'], console_type_sales['Units sold (million)'], color='skyblue')
plt.title('Total Units Sold by Console Type')
plt.xlabel('Console Type')
plt.ylabel('Total Units Sold (million)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
